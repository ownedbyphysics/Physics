```python
import numpy as np
import matplotlib.pyplot as plt

# Number of particles and physical constants
num_particles = 10000
temperature = 300  # in Kelvin
mass = 4.65e-26    # mass of a nitrogen molecule in kg
k_B = 1.38e-23     # Boltzmann constant in J/K

# Compute standard deviation for the velocity distribution
sigma = np.sqrt(k_B * temperature / mass)

# Generate random velocities in 3 dimensions using Gaussian distribution
v_x = np.random.normal(0, sigma, num_particles)
v_y = np.random.normal(0, sigma, num_particles)
v_z = np.random.normal(0, sigma, num_particles)

# Calculate the magnitude of velocity for each particle
v = np.sqrt(v_x**2 + v_y**2 + v_z**2)

# Plot histogram of simulated velocity magnitudes
plt.hist(v, bins=50, density=True, alpha=0.6, label="Simulated data")

# Compute the theoretical MaxwellBoltzmann distribution manually
v_space = np.linspace(0, np.max(v), 500)
factor = 4 * np.pi * (mass / (2 * np.pi * k_B * temperature))**(3/2)
exponential_term = np.exp(-mass * v_space**2 / (2 * k_B * temperature))
theoretical_mb = factor * v_space**2 * exponential_term

# Plot the theoretical MaxwellBoltzmann distribution
plt.plot(v_space, theoretical_mb, 'r-', label="Theoretical MaxwellBoltzmann")

# Plot configuration
plt.xlabel("Speed (m/s)")
plt.ylabel("Probability Density")
plt.title("Particle Speed Distribution (MaxwellBoltzmann)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
