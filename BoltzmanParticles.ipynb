```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import maxwell

# Αριθμός σωματιδίων και φυσικές σταθερές
num_particles = 10000
temperature = 300  # σε Kelvin
mass = 4.65e-26    # μάζα μορίου (π.χ. άζωτο) σε kg
k_B = 1.38e-23     # σταθερά Boltzmann

# Υπολογισμός της τυπικής απόκλισης (συντελεστής κλίμακας Maxwell)
sigma = np.sqrt(k_B * temperature / mass)

# Δημιουργία τυχαίων ταχυτήτων σε 3 διαστάσεις
v_x = np.random.normal(0, sigma, num_particles)
v_y = np.random.normal(0, sigma, num_particles)
v_z = np.random.normal(0, sigma, num_particles)

# Υπολογισμός μέτρου ταχύτητας
v = np.sqrt(v_x**2 + v_y**2 + v_z**2)

# Ιστόγραμμα των ταχυτήτων
plt.hist(v, bins=50, density=True, alpha=0.6, label="Προσομοιωμένα δεδομένα")

# Θεωρητική καμπύλη Maxwell–Boltzmann
v_space = np.linspace(0, np.max(v), 500)
mb_pdf = maxwell.pdf(v_space, scale=sigma)
plt.plot(v_space, mb_pdf, 'r-', label="Θεωρητική κατανομή Maxwell–Boltzmann")

# Ρυθμίσεις γραφήματος
plt.xlabel("Ταχύτητα (m/s)")
plt.ylabel("Πυκνότητα πιθανότητας")
plt.title("Κατανομή ταχυτήτων σωματιδίων (Maxwell–Boltzmann)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
