```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import maxwell

#     
num_particles = 10000
temperature = 300  #  Kelvin
mass = 4.65e-26    #   (.. )  kg
k_B = 1.38e-23     #  Boltzmann

#     (  Maxwell)
sigma = np.sqrt(k_B * temperature / mass)

#     3 
v_x = np.random.normal(0, sigma, num_particles)
v_y = np.random.normal(0, sigma, num_particles)
v_z = np.random.normal(0, sigma, num_particles)

#   
v = np.sqrt(v_x**2 + v_y**2 + v_z**2)

#   
plt.hist(v, bins=50, density=True, alpha=0.6, label=" ")

#   MaxwellBoltzmann
v_space = np.linspace(0, np.max(v), 500)
mb_pdf = maxwell.pdf(v_space, scale=sigma)
plt.plot(v_space, mb_pdf, 'r-', label="  MaxwellBoltzmann")

#  
plt.xlabel(" (m/s)")
plt.ylabel(" ")
plt.title("   (MaxwellBoltzmann)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
